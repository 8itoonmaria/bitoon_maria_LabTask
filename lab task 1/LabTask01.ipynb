### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
# Gender to numeric
df['Sex_numeric'] = df['Sex'].map({'male': 0, 'female': 1})
```

creates a new column and replaces values based on predefined dictionary using map()

```python
# Age groups
def categorize_age(age):
    if pd.isna(age):
        return 'unknown'
    elif age < 18:
        return 'child'
    elif age < 60:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(categorize_age)
```
